```python
import pandas as pd
import bamboolib as bam
df = pd.read_csv("../data/titanic.csv")
Interval = pd.Interval
```

```python
df = df.drop(columns=['Fare', 'Survived', 'Pclass', 'Name', 'Age', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'])
df.dtypes
```

```python
df['PassengerId_Integer'] = pd.to_numeric(df['PassengerId'], downcast='integer', errors='coerce')
df['PassengerId_String'] = df['PassengerId'].astype('string')
df['PassengerId_Float'] = pd.to_numeric(df['PassengerId'], downcast='float', errors='coerce')
df['PassengerId_Boolean'] = df['PassengerId'].astype(bool)
# df['PassengerId_Categorical'] = df['PassengerId'].astype('category')
df['PassengerId_Timedelta'] = pd.to_timedelta(df['PassengerId'])
df['PassengerId_Object'] = df['PassengerId'].astype('object')
df['PassengerId_bins'] = pd.cut(df['PassengerId'], bins=10)
df['PassengerId_bins_string_category'] = pd.cut(df['PassengerId'], bins=10).astype('string').astype('category')
df['PassengerId_Int32'] = df['PassengerId'].astype('Int32')
df['PassengerId_Uint32'] = df['PassengerId'].astype('uint32')
df['PassengerId_binarized'] = (df['PassengerId'] % 2 == 0)*1
df['PassengerId_binarized_c'] = pd.to_numeric(df['PassengerId_binarized'], downcast='integer', errors='coerce')
df['PassengerId_binarized_c_bool'] = df['PassengerId_binarized_c'].astype(bool)
df['PassengerId_binarized_c_category'] = df['PassengerId_binarized_c'].astype('category')
df['Sex_Category'] =  df['Sex'].astype('category')
print(df.dtypes)
df2 = df.loc[df['Sex_Category'] == 'male']
# df2 = df.loc[df['PassengerId_bins_string_category'] == '(0.11, 90.0]' ]
df2 = df2.loc[df2['Sex_Category'].isin(['female'])]
df2['PassengerId_bins_string_category'] = df2['PassengerId_bins_string_category'].astype('string')
df
```

```python
split_df = df2["PassengerId_bins_string_category"].str.split(',', expand=True)
split_df.columns = [f"PassengerId_bins_string_category_{id_}" for id_ in range(len(split_df.columns))]
df2 = pd.merge(df2, split_df, how="left", left_index=True, right_index=True)
df2 = df2.reset_index()
df2
```
